# 직전값과 직후값의 평균으로 결측치 처리 

직전값과 직후값의 평균으로 fillna 하는 함수 

```python
# 결측치 처리하는 함수

def neighbor_fillna(dataframe, nan_list, column_name):
    filler = dict()
    for i in nan_list:
        before = i-1
        after = i+1
        if df.loc[before][column_name]!=df.loc[before][column_name]:
            # 직후값이 결측치인 경우 한칸 더 아래로 가기 
            before = before - 1
        if df.loc[after][column_name]!=df.loc[after][column_name]:
            # 직후값이 결측치인 경우 한칸 더 아래로 가기 
            after = after + 1
        # 결측치의 앞뒤 값 평균 계산  
        fill_value = (dataframe.loc[before][column_name] + dataframe.loc[after][column_name]) / 2
        # dict에 추가 
        filler[i]=fill_value
    print(filler)
    # 결측치 채우기
    dataframe[column_name] = dataframe[column_name].fillna(filler)
    return dataframe
```

기온 결측치 처리 

```python
col_name = '기온(°C)'
# 결측치 index를 리스트로 받기 
nanlist = list(df[df[col_name].isna()==True].index)
# 앞뒤 값 평균으로 결측치 처리 
df = neighbor_fillna(df,nanlist,col_name)
```

```python
# 잘 처리되었는지 확인
df.isna().sum()
```

습도 결측치 처리 후 int 변환
